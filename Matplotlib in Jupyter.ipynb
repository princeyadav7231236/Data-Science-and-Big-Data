```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1, 2, 3, 4, 5, 6, 7, 8, 9], [1, 2, 13, 4, 5, 16, 17 , 8, 19])
plt.title("This is my first matplotlib chart")
plt.xlabel("These are values along X axis")
plt.ylabel("These are values along Y axis")
plt.show()
```
